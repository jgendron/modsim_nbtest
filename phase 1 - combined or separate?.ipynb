```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_all = pd.read_table('./data/all208_composition.txt',sep='\t',
                       header=None, index_col=0)
df_all[1] = df_all[1].apply(lambda x: x[54:61])
df_all.columns= list(range(-1,df_all.shape[1]-1))
df_all.set_index(-1, inplace=True)
df_all.sort_values(by=-1, inplace=True)
df_all.head(12)
```

```python
df_combined = pd.read_table('./data/allCombined_composition.txt',sep='\t',
                       header=None, index_col=0)
df_combined[1] = df_combined[1].apply(lambda x: x[63:70])
df_combined.columns= list(range(-1,df_combined.shape[1]-1))
df_combined.set_index(-1, inplace=True)
df_combined.sort_values(by=-1, inplace=True)
df_combined.head(12)
```

```python
cmap = sns.choose_cubehelix_palette(as_cmap=False)
cmap = sns.choose_light_palette(as_cmap=False)
```

```python
sns.clustermap(df_all,row_cluster=False,col_cluster=True, cmap=cmap )
sns.heatmap
```

```python
sns.clustermap(df_combined,row_cluster=False,col_cluster=False, cmap=cmap )
sns.heatmap
```

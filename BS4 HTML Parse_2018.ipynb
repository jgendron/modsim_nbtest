```python
import requests
r = requests.get('http://modsimworld.org/conference-papers/2018')
```

```python
print(r.text)
```

```python
from bs4 import BeautifulSoup 
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
results = soup.find_all('table')
```

```python
len(results)
```

```python
first_result = results[0]
```

```python
first_result
```

```python
#Extract Track 
first_result.find('th').text
```

```python
#Extract Title 
first_result.find('span').text
```

```python
#Extract Author 
first_result.find('td').text
```

```python
#Extract Filename 
first_result.find('a')['href']
```

```python
#Extract Year 
first_result.find('a')['href'][8:12]
```

```python
Records = [] 
for result in results: 
    Track = result.find('th').text
    Filename = result.find('a')['href']
    Author = result.find('td').text
    Title = result.find('span').text 
    Year = result.find('a')['href'][8:12]
    Records.append((Track, Filename, Author, Title, Year))
    
   
```
